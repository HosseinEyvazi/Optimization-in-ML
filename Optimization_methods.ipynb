# Understanding Gradient Descent: A Short Guide

## What is Gradient Descent?

Gradient Descent is an optimization algorithm used to minimize a function by iteratively moving in the direction of steepest descent (negative gradient). It is widely used in machine learning, especially in training models like linear and logistic regression.

## Example of Gradient Descent

Suppose we want to find the minimum of the function:

\[
f(x) = (x - 3)^2
\]

The derivative (gradient) is:

\[
f'(x) = 2(x - 3)
\]

We update \( x \) using the rule:

\[
x_{\text{new}} = x_{\text{old}} - \alpha f'(x)
\]

If we start at \( x = 10 \) and choose a learning rate, the sequence will converge towards \( x = 3 \), the minimum point.

## Key Hyperparameters and How to Choose Them

### Learning Rate (\(\alpha\))  Determines step size.

- **Too large**: May overshoot the minimum.
- **Too small**: Converges slowly.

**Tip:** Start with 0.01 or 0.1 and adjust based on performance.

### Batch Size (for mini-batch descent)

- **Small batch**: Noisy but can escape local minima.
- **Large batch**: Stable but may get stuck.

**Tip:** Common values are **32, 64, or 128**.

### Stopping Criteria

- When gradient is very small (\(< \epsilon\)).
- When cost function change is negligible.

**Tip:** Set a tolerance like \( 10^{-6} \).

## Why is it Called "Gradient Descent"?

The term **descent** is used because the algorithm moves in the **negative direction** of the gradient, reducing the function value in each step, leading towards the lowest point (minimum).

## Gradient Vector and Derivatives

The **gradient vector** is a collection of all partial derivatives of a function.

If `f(x, y)` is a function, then:

## When Does Gradient Descent Stop?

Gradient descent stops when:

- The gradient is close to zero.
- The cost function no longer decreases significantly.
- A maximum number of iterations is reached.

By tuning hyperparameters wisely and using proper stopping criteria, gradient descent efficiently finds optimal solutions in machine learning models.
